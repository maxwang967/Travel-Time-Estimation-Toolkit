```python
import folium
import pandas as pd
```

```python
# 1 pre-load matched-edges file to obtain mapping (fid -> geometry)
def extract_list_of_points_for_matched_edges(mgeom):
    # print(mgeom)
    line_str = mgeom.replace("LINESTRING (", "[(").replace(")", ")]").replace(", ", "),(").replace(" ", ",")
    points = eval(line_str)
    if len(points[0]) == 0:
        return []
    points = list(map(lambda x: [x[1], x[0]], points))
    return points
matched_edges_file = "../data/didi/matched_edges.csv"
df_me = pd.read_csv(matched_edges_file, sep=";")
df_me.dropna(how="any", inplace=True)
df_me["folium_lines"] = df_me.apply(lambda row: extract_list_of_points_for_matched_edges(row.geometry), axis=1)
df_me
```

```python
# 2 load matched routes from mr.txt
def extract_list_of_points_for_mr(mgeom):
    # print(mgeom)
    line_str = mgeom.replace("LINESTRING(", "[(").replace(")", ")]").replace(",", "),(").replace(" ", ",")
    points = eval(line_str)
    if len(points[0]) == 0:
        return []
    points = list(map(lambda x: [x[1], x[0]], points))
    return points
mr_file = "../data/didi/mr.txt"
df_mr = pd.read_csv(mr_file, sep=";")
df_mr.dropna(how="any", inplace=True)
df_mr["folium_lines"] = df_mr.apply(lambda row: extract_list_of_points_for_mr(row.mgeom), axis=1)
df_mr
```

```python
# 3 convert list of routes (cpath using fid) to list of geometry
def extract_list_of_converted_geometry(cpath):
    routes_to_geometry = []
    for fid in cpath.split(","):
        # print(fid)
        routes_to_geometry += df_me.loc[df_me.fid == int(fid)].folium_lines.values[0]
    # print(routes_to_geometry
    return routes_to_geometry
df_mr["route_folium_lines"] = df_mr.apply(lambda row: extract_list_of_converted_geometry(row.cpath), axis=1)
df_mr.to_pickle("../data/didi/df_mr.pkl")
```
