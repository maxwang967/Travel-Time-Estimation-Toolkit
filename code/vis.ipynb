```python
import folium
import pandas as pd
```

```python
mr_file = "../data/mr.txt"
df = pd.read_csv(mr_file, sep=";")
df.dropna(how="any", inplace=True)
```

```python
df
```

```python
def extract_list_of_points(mgeom):
    # print(mgeom)
    line_str = mgeom.replace("LINESTRING(", "[(").replace(")", ")]").replace(",", "),(").replace(" ", ",")
    points = eval(line_str)
    if len(points[0]) == 0:
        return []
    points = list(map(lambda x: [x[1], x[0]], points))
    return points
```

```python
df["folium_lines"] = df.apply(lambda row: extract_list_of_points(row.mgeom), axis=1)
```

```python
len(df[df.mgeom == "LINESTRING()"]) / len(df)
```

```python
df
```

```python
m = folium.Map(location=[41.1459905, -8.6122327])
```

```python
points_list = df.folium_lines[:100]
for pl in points_list:
    folium.Marker(pl[0]).add_to(m)
    folium.Marker(pl[-1], icon=folium.Icon(color="green")).add_to(m)
    folium.PolyLine(pl).add_to(m)
    
```
